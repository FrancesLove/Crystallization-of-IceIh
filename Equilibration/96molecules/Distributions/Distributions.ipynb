# Calculate distributions of the kernel for liquid water and ice Ih

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
###################################################################
# Plot options
###################################################################

font = {'family' : 'serif',
        'serif'   : 'palatino',
        #'sans-serif'    : 'Computer Modern Sans serif', 
        'style'   : 'normal',
        'variant'   : 'normal',
        'stretch'   : 'normal',
        'weight'   : 'normal',
        'size'   : 20}
plt.rc('font', **font)
plt.rc('text', usetex=True)
plt.rcParams['figure.figsize'] = (15, 6)
```

```python
dataSolid=np.genfromtxt("histoSolid")
dataLiquid=np.genfromtxt("histoLiquid")
```

Plot distributions

```python
plt.fill_between(dataLiquid[:,0],0,dataLiquid[:,1],alpha=0.2)
plt.fill_between(dataSolid[:,0],0,dataSolid[:,1],alpha=0.2)
plt.plot(dataLiquid[:,0],dataLiquid[:,1],label="Liquid")
plt.plot(dataSolid[:,0],dataSolid[:,1],label="Ice Ih")

plt.xlim([0,1])
plt.ylim([0,10])
plt.xlabel(r"$k_{\chi_0}(\chi)$")
plt.ylabel("Probability")
plt.yticks([0,5,10])
plt.legend()

plt.savefig("distributions.png", bbox_inches='tight')

plt.show()
```
